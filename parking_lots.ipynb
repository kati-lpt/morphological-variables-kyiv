```python
import pandas as pd
import geopandas as gpd
```

```python
blocks = gpd.read_file('blocks.geojson')[['study_case','area','id', 'geometry']]
blocks
# Import blocks file - cosists geometry, microrayon name and block area
```

```python
parking = gpd.read_file('parking_lots.geojson')[['capacity', 'geometry']]
parking.capacity = parking.capacity.astype('int')
parking
```

```python
blocks\
    .sjoin(parking, predicate='contains')\
    .drop(columns=['geometry', 'index_right'])\
    .groupby(['id','study_case','area']).sum()
```
