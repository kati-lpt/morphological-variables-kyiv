```python
import pandas as pd
import geopandas as gpd
```

```python
blocks = gpd.read_file('blocks.geojson')
blocks = blocks.set_index('id').sort_values(by='id')
blocks
# Import blocks file - cosists geometry, microrayon name and block area
```

```python
buildings = gpd.read_file('buildings.geojson') 
#  Import buildings file - consists type of building, area, study case, OSM info, number of apartments, building type and geometry.
buildings
```

```python
buildings.groupby('building')[[]].count()
```

# dwelling unit mix:
### single family:
- house

### multi-family:
- apartments
- residential

### mixed-use residential:
- apartments & mix-used
- residential & mix-used
- house & mix-used

```python
features=gpd.read_file("nodes_names.geojson")
features
```

```python
mix_use = features.sjoin(buildings, predicate="within", how="inner").groupby("index_right").count()[[]]
mix_use['use'] = 'mixed use'
mix_use
```

```python
buildings = buildings.join(mix_use)
buildings
```

```python
residential = buildings[(buildings.building=="apartments") | (buildings.building=="house") | (buildings.building=="residential")]
residential
```

```python
residential.loc[((residential.building=='house') & (residential.use!='mixed use')), 'use' ]='single family'
residential.loc[(((residential.building=='residential') | (residential.building=='apartments')) & (residential.use!='mixed use')), 'use' ]='multi family'
residential
```

```python
#residential.to_file('dwelling unit mix.geojson')
```

```python
blocks['mix_use_residential_number'] = residential[residential.use=='mixed use'].groupby('block_id').count()['use']
blocks['single_family_number'] = residential[residential.use=='single family'].groupby('block_id').count()['use']
blocks['multi_family_number'] = residential[residential.use=='multi family'].groupby('block_id').count()['use']
blocks['total_residential'] = residential.groupby('block_id').count()['use']
blocks
```

```python
blocks['mix_use_residential_percent'] = blocks.mix_use_residential_number / blocks.total_residential
blocks['single_family_percent'] = blocks.single_family_number / blocks.total_residential
blocks['multi_family_percent'] = blocks.multi_family_number / blocks.total_residential
blocks
```
